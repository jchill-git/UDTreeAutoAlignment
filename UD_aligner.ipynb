```python
from collections import Counter
import math

def counter_cosine_similarity(c1, c2):
    terms = set(c1).union(c2)
    dotprod = sum(c1.get(k, 0) * c2.get(k, 0) for k in terms)
    magA = math.sqrt(sum(c1.get(k, 0)**2 for k in terms))
    magB = math.sqrt(sum(c2.get(k, 0)**2 for k in terms))
    return dotprod / (magA * magB)
```

```python
sim_score=0
for sent in tqdm(trans_UD.keys()):

    gk_pop=list()
    for word in sentence_UD[sent]:
        gk_pop.append(word[2])

    trans_pop=list()
    for word in trans_UD[sent]:
        trans_pop.append(word[2].lower())

    sim_score=sim_score+counter_cosine_similarity(Counter(gk_pop),Counter(trans_pop))
sim_score/len(trans_UD.keys())
```
